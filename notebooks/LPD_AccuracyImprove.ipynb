```python
import cv2
import numpy as np
import torch
from ultralytics import YOLO
import pytesseract
import easyocr
import re
import matplotlib.pyplot as plt
from fuzzywuzzy import process
```

```python
model = YOLO("runs/detect/train2/weights/best.pt")
```

```python
def detect_license_plate(image_path):
    results = model.predict(source=image_path, conf=0.5)  # Get predictions
    img = cv2.imread(image_path)

    for result in results:
        for box in result.boxes:
            # Get bounding box coordinates
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            cropped_plate = img[y1:y2, x1:x2]  # Crop license plate

            # Draw bounding box on the original image
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)

            return cropped_plate, img  # Return cropped plate & annotated image
    return None, img
```

```python
def preprocess_plate(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
    gray = cv2.GaussianBlur(gray, (3, 3), 0)  # Apply blur to reduce noise

    # Adaptive thresholding for uneven lighting
    thresh = cv2.adaptiveThreshold(
        gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2
    )

    # Morphological closing to enhance characters
    kernel = np.ones((2, 2), np.uint8)
    processed = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)

    return processed
```

```python
def extract_text_easyocr(plate):
    reader = easyocr.Reader(['en'])  # Initialize EasyOCR
    results = reader.readtext(plate, detail=0)  # Extract text
    text = " ".join(results) if results else "N/A"

    # Clean text to keep only uppercase letters and numbers
    clean_text = "".join(re.findall(r"[A-Z0-9\s]", text)).strip()

    return clean_text
```

```python
def extract_text_tesseract(plate):
    text = pytesseract.image_to_string(
        plate,
        config="--psm 7 -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789"
    )
    return text.strip()
```

```python
def validate_license(text):
    # Regex for Indian license plate format
    pattern = r'^(?:IND\s?)?[A-Z]{2}\s?\d{2}\s?[A-Z]{1,3}\s?\d{4}$'
    is_valid = bool(re.match(pattern, text))

    if is_valid:
        if text.startswith("IND"):
            return "âœ… Valid License Plate"
        else:
            return "âœ… Valid License Plate (Consider adding 'IND')"
    return "âŒ Invalid License Plate"
```

```python
VALID_STATE_CODES = [
    "MH", "DL", "KA", "TN", "GJ", "UP", "RJ", "WB",
    "MP", "PN", "HP", "CH", "BR", "AR", "AP", "AS",
    "GA", "DN", "HR", "JK", "KL", "LD", "MN", "MZ",
    "NL", "ML", "OR", "SK", "TR"
]

def correct_state_code(text):
    words = text.split()
    if len(words) > 0:
        best_match, score = process.extractOne(words[0], VALID_STATE_CODES)
        if score > 80:  # If high confidence, replace it
            words[0] = best_match
    return " ".join(words)
```

```python
def main(image_path):
    # Step 1: Detect the license plate
    plate, detected_image = detect_license_plate(image_path)

    if plate is not None:
        # Step 2: Preprocess the plate
        processed_plate = preprocess_plate(plate)

        # Step 3: Try EasyOCR, fallback to Tesseract if needed
        plate_number = extract_text_easyocr(processed_plate)
        if len(plate_number) < 6:  # If result too short, use Tesseract
            plate_number = extract_text_tesseract(processed_plate)

        # Step 4: Correct state code
        plate_number = correct_state_code(plate_number)

        # Step 5: Validate the plate format
        validation_msg = validate_license(plate_number)

        # Step 6: Display the results
        plt.figure(figsize=(10, 5))

        # Show detected license plate on the image
        plt.subplot(1, 2, 1)
        plt.imshow(cv2.cvtColor(detected_image, cv2.COLOR_BGR2RGB))
        plt.title("Detected License Plate")

        # Show preprocessed plate and extracted text
        plt.subplot(1, 2, 2)
        plt.imshow(processed_plate, cmap='gray')
        plt.title(f"OCR: {plate_number}\n{validation_msg}")

        plt.show()
    else:
        print("No license plate detected.")
```
